# English Dictionary

## This is a desktop based dictionary in which you can add any word and their meanings

## and can be searched later.

```python
from tkinter import *
from tkinter import messagebox

# creating main window

win=Tk()

win.state("zoomed")
win.configure(bg="#FFBB70")
win.title("English Dictionary")

#Creating main label

main_label=Label(win,text="English Dictionary",font=("algerian",50,'bold'),bg="#FFBB70",pady=30)
main_label.pack()

#Creating first frame for Adding any word

add_frm=Frame(win,highlightthickness=2,highlightbackground='#ED9455')
add_frm.configure(bg="#FFEC9E")
add_frm.place(x=50,y=150,width=590,height=570)


def store():
    word=input_entry.get()
    mean=input_mean.get(1.0,'end')
    try:
        import pymongo
        con=pymongo.MongoClient()
        db=con['dictionary']
        col=db['wordmean']
        col.insert_one({'word':word,'mean':mean});
        con.close()
        messagebox.showinfo("Successfull","Word Added successfully")
        input_entry.delete(0,'end')
        input_mean.delete(1.0,'end')
        
    except:
        messagebox.showinfo("Unsuccessfull","Word couldn't Added")
        
def reset():
    input_entry.delete(0,'end')
    input_mean.delete(1.0,'end')
    
    
#creating Label for frame 1

add_label=Label(add_frm,text="Add Word and Meaning",font=("cambria",30,'bold'),bg="#ED9455")
add_label.pack()

input_label=Label(add_frm,text="Input Word :",font=("cambria",20,'bold'),bg="#FFEC9E")
input_label.place(x=10,y=100)

input_entry=Entry(add_frm,font=("cambria",20,'bold'),bd=2)
input_entry.place(x=220,y=100)
input_entry.focus()

input_mean_label=Label(add_frm,text="Input Meaning :",font=("cambria",20,'bold'),bg="#FFEC9E")
input_mean_label.place(x=10,y=193)

input_mean=Text(add_frm,font=("cambria",20,'bold'),bd=2)
input_mean.place(x=220,y=200,width=327,height=200)


input_button=Button(add_frm,text="Add Word",font=("cambria",20,'bold'),bd=2,command=store)
input_button.place(x=310,y=410)

reset_button=Button(add_frm,text="Reset",font=("cambria",20,'bold'),bd=2,command=reset)
reset_button.place(x=340,y=470)



#frame 1 finish here---------------------------------------------------------------------------------------------------------


#Creating second frame for searching any word-----------------------------------------------------------------------------------

search_frm=Frame(win,highlightthickness=2,highlightbackground='#ED9455')
search_frm.configure(bg="#FFEC9E")
search_frm.place(x=700,y=150,width=610,height=570)



def search():
    global word1
    word1=search_entry.get()
    li=[]
    try:
        
        import pymongo
        con=pymongo.MongoClient()
        db=con['dictionary']
        col=db['wordmean']
        mean=col.find({'word':word1})
        for i in mean:
            li.append(i)
        con.close()
        output_text.insert(1.0,li[0]['mean'])
    except:
        messagebox.showinfo("Unsuccessfull","Word Not Found")
        
        

def update():
    word=search_entry.get()
    mean=output_text.get(1.0,'end-1c')
    try:
        import pymongo
        con=pymongo.MongoClient()
        db=con['dictionary']
        col=db['wordmean']
        find=({'word':word1})
        new=({"$set":{'word':word,'mean':mean}})
        col.update_one(find,new)
        con.close()
        messagebox.showinfo("Successfull","Word Updated Successfully")
    except:
        messagebox.showinfo("Unsuccessfull","Word Couldn't Update")
        

def reset_out():
    search_entry.delete(0,'end')
    output_text.delete(1.0,'end')
    
def remove():
    word=search_entry.get()
    try:
        import pymongo
        con=pymongo.MongoClient()
        db=con['dictionary']
        col=db['wordmean']
        col.delete_one({'word':word})
        con.close()
        messagebox.showinfo("Successfull","Removed Successfully")
        search_entry.delete(0,'end')
        output_text.delete(1.0,'end')
    except:
        messagebox.showinfo("Unsuccessfull","Not Removed")
    
#creating Label for frame 2

search_main_label=Label(search_frm,text="Search Word and Meaning",font=("cambria",30,'bold'),bg="#ED9455")
search_main_label.pack()

search_label=Label(search_frm,text="Search Word :",font=("cambria",20,'bold'),bg="#FFEC9E")
search_label.place(x=10,y=100)

search_entry=Entry(search_frm,font=("cambria",20,'bold'),bd=2)
search_entry.place(x=240,y=100)

output_mean_label=Label(search_frm,text="Word's Meaning :",font=("cambria",20,'bold'),bg="#FFEC9E")
output_mean_label.place(x=10,y=193)

output_text=Text(search_frm,font=("cambria",20,'bold'),bd=2)
output_text.place(x=240,y=200,width=327,height=200)


output_button=Button(search_frm,text="Search",font=("cambria",20,'bold'),bd=2,command=search)
output_button.place(x=320,y=410)

update_button=Button(search_frm,text="Update",font=("cambria",20,'bold'),bd=2,command=update)
update_button.place(x=455,y=470)

reset_search_button=Button(search_frm,text="Reset",font=("cambria",20,'bold'),bd=2,command=reset_out)
reset_search_button.place(x=464,y=410)

remove_button=Button(search_frm,text="Remove",font=("cambria",20,'bold'),bd=2,command=remove)
remove_button.place(x=312,y=470)

win.mainloop()
```
